```python
# instead of categorising each column at a time
# better way is to create user defined functions

drugy_codes = {'never':0, 'sometimes':1, 'often':2}
smokey_codes = {'sometimes':0, 'when drinking':0, 'yes':0, 'trying to quit':0, 'no':1}
drinky_codes = {'socially':0, 'often':1, 'not at all':2, 'rarely':3, 'very often':4, 'desperately':5}
body_type_codes = {'thin':0, 'skinny':1, 'fit':2, 'athletic':3, 'jacked':4, 'rather not say':5, 'average':6, 'a little extra':7, 'used up':8, 'curvy':9, 'overweight':10, 'fully figured':11}
```

```python
#create function to convert categorical to numeric
def categorise(df):

  # categorising various features to numeric
  df['drugy_codes'] = df['drugs'].map(drugy_codes)
  df['smokey_codes'] = df['smokes'].map(smokey_codes)
  df['drinky_codes'] = df['drinks'].map(drinky_codes)

  return
```

```python
# copy the original dataframe
df_copy = df
# calling the function
categorise(df_copy)
```

```python
print("training_accuracy:", training_accuracy)
print(classification_report(y_test, predictions))
```

```python
# Model 4: Logistic Regression
from sklearn.linear_model import LogisticRegression
```
